```python
import requests
from bs4 import BeautifulSoup
import collections
from collections import Counter

count = Counter()

for skill in ["python","Java","MySQL"]:
        url = "http://www.104.com.tw/i/apis/jobsearch.cfm?kws={}&kwop=3&area=6001001000&order=2&fmt=4&cols=J,JOB,NAME".format(skill)
        job = requests.get(url)
        soup = BeautifulSoup(job.text,'lxml')
        for page in range(1,int(soup.list['totalpage'])):
            url = "http://www.104.com.tw/i/apis/jobsearch.cfm?kws={}&kwop=3&area=6001001000&order=2&fmt=4&page={}&cols=J,JOB,NAME".format(skill,page)
            job = requests.get(url)
            soup = BeautifulSoup(job.text,'lxml')
            joblist = soup.find_all('item')
            for jobitem in joblist:
                if jobitem in count:
                    count[jobitem['job']]+=1
                else:
                    count[jobitem['job']]=1    

count.most_common()
```
